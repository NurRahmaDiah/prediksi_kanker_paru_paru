# RATA-RATA UMUR YANG TERKENA KANKER PARU-PARU

```python
rerata = df.groupby('AGE').count()[['LUNG_CANCER']].sort_values(by='LUNG_CANCER',ascending=True).reset_index()
rerata = rerata.rename(columns={'LUNG_CANCER':'FIX KANKER'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=rerata['AGE'], y=rerata['FIX KANKER'], color='yellow')
plt.xticks(rotation=60)
```

# RATA-RATA PEROKOK YANG TERKENA KANKER PARU-PARU

```python
roko = df.groupby('LUNG_CANCER').count()[['SMOKING']].sort_values(by='SMOKING',ascending=True).reset_index()
roko = roko.rename(columns={'SMOKING':'FIX KANKER','LUNG_CANCER' : 'MEROKOK 1 YES 2 NO'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=roko['MEROKOK 1 YES 2 NO'], y=roko['FIX KANKER'], color='yellow')
plt.xticks(rotation=60)
```
